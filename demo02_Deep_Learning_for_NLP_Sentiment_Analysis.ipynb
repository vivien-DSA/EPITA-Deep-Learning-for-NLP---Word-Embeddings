## Q1 : Use Tokenizer from keras to process data and labels

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# processing data
texts = pre_processed_tweets_list
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
data = pad_sequences(sequences, maxlen=32)
vocab = tokenizer.word_index
vocab['<eos>'] = 0
```

```python
# processing labels 
# the label are the 5th column of the data 
texts_label = csv_file_data.iloc[:, 4]
tokenizer_label = Tokenizer(num_words=6)
tokenizer_label.fit_on_texts(texts_label)
sequences_label = tokenizer_label.texts_to_sequences(texts_label)
labels = pad_sequences(sequences_label, maxlen=1)
vocab_labels = tokenizer_label.word_index
vocab_labels['<eos>'] = 0
```

```python
import numpy as np
from keras.utils import to_categorical
# one hot encode
encoded = to_categorical(labels)
# take just the 4 last column of the encoded as labels
labels = encoded[:, 1:5]
```

```python
from sklearn.utils import shuffle

data, labels = shuffle(data, labels)

x_train, y_train = data[:4000], labels[:4000]
x_val, y_val = data[4000:], labels[4000:]
```

# II. Train and evaluate the model

```python
from tp_sentiment_analysis import embedding

weights = embedding.load(vocab, 100, \
    'tp_sentiment_analysis/glove.twitter.27B.100d.filtered.txt')
pd.DataFrame(weights)
```

```python
vocab.keys()
```

```python
y_train[0]
```

```python
from keras.layers import Embedding, Input, GRU, Dense
from keras.models import Model
import tensorflow as tf
# 100-dim embeddings initialized with GloVe,
# over sequences of size 32, and not fine tuneable
embedding_layer = Embedding(len(vocab), 100, weights=[weights],input_length=32, trainable=False)
sequence_input = Input(shape=(32,), dtype='int32')
embedded_sequences = embedding_layer(sequence_input)
x = GRU(64)(embedded_sequences)
preds = Dense(labels.shape[1], activation='softmax')(x)
model_GRU = Model(sequence_input, preds)
model_GRU.compile(loss='categorical_crossentropy',optimizer='Nadam', metrics=['acc'])
print(model_GRU.summary())
model_GRU.fit(x_train, y_train,\
          validation_data=(x_val, y_val),\
          epochs=20, batch_size=64, shuffle=True)
```

## Q2 : Evaluate the model using precision, recall, F1-measure and accuracy metrics

To do so, we add those metrics in the configuration of the compiler of our model as follow:

```python
import tensorflow as tf
import tensorflow_addons as tfa

model_GRU.compile(loss='categorical_crossentropy',optimizer='Nadam',\
              metrics=[tf.keras.metrics.Precision(),\
                       tf.keras.metrics.Recall() ,\
                       tfa.metrics.F1Score(num_classes=4, average='macro'),\
                       'acc'])
model_GRU.fit(x_train, y_train,\
          validation_data=(x_val, y_val),\
          epochs=20, batch_size=64, shuffle=True)
```

```python
model_GRU.metrics_names
```

```python
loss, precision, recall, f1_measure, acc =\
                model_GRU.evaluate(x_train, y_train, batch_size=64)

print("Test loss :", loss)
print("Test precision:", precision)
print("Test recall:", recall)
print("Test F1-measure:", f1_measure)
print("Test accuracy:", acc)
```

```python
from keras.layers import Conv1D, MaxPooling1D, Flatten
# We need an Input to use Model() method of keras.models
sequence_input = Input(shape=(32,), dtype='int32')
# We create the first layer of the GRU model
embedded_sequences = embedding_layer(sequence_input)
# Now we apply the layers of CNN 
x = Conv1D(128, 3, activation='relu')(embedded_sequences)
x = MaxPooling1D(3)(x) # global max pooling
x = Flatten()(x)
# Final layer giving output with softmax probability 
preds = Dense(labels.shape[1], activation='softmax')(x)
# Ne can create our model with the input and the output
model_CNN = Model(sequence_input, preds)
# We specify the configuration (loss, metrics, etc.)
model_CNN.compile(loss='categorical_crossentropy',\
                  optimizer='Nadam',\
                  metrics=['acc'])
# We train the model
model_CNN.fit(x_train, y_train, \
              validation_data=(x_val, y_val),\
              epochs=10, batch_size=64, shuffle=True)
```
